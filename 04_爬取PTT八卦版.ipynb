## 載入套件

```python
import requests
from bs4 import BeautifulSoup
```

## 取得頁面資料


```python
url = 'https://www.ptt.cc/bbs/Gossiping/index.html'

resp = requests.get(url)
soup = BeautifulSoup(resp.text, 'lxml')
soup
```

# 帶入 cookie
透過帶入 cookie 讓對方伺服器認為我們已經點過『我同意』了

```python
resp = requests.get(url, cookies={'over18': '1'})
soup = BeautifulSoup(resp.text, 'lxml')
soup
```

# 取出文章

```python
for item in soup.select('.r-ent'):
    title = item.select('.title')[0].text
    like = item.select('.nrec span')[0].text
    author = item.select('.author')[0].text
    date = item.select('.date')[0].text
    link = item.select('.title a')[0]['href']
    print('title = ', title)
    print('like = ', like)
    print('author = ', author)
    print('date = ', date)
    print('link = ', link)
    print('========')
```

# 抓取上一頁小範例

```python
# 把取資料的內容寫成 func 方便重複呼叫
def get_page(url):
    hostname = 'https://www.ptt.cc'
    resp = requests.get(url, cookies={'over18': '1'})
    soup = BeautifulSoup(resp.text, 'lxml')
    prev_page_url = hostname + soup.select('.btn-group-paging a')[1]['href']
    return soup, prev_page_url
```

```python
# 取得前五頁的連結
url = 'https://www.ptt.cc/bbs/Gossiping/index.html'
for i in range(5):
    current_page, prev_page_url = get_page(url)
    print(prev_page_url)
    url = prev_page_url
```

# 取得推數超過30的文章
[PTT 網頁版也可以使用站內搜尋功能](https://disp.cc/b/163-aAKM)

搜尋推數超過30的文章網址: https://www.ptt.cc/bbs/Gossiping/search?q=recommend%3A30

網址最面的數字30代表取得超過多少推數的文章，我們可以自己去更改數字取得不同推數的文章

```python
like_num = 30
url = 'https://www.ptt.cc/bbs/Gossiping/search?page=1&q=recommend%3A{}'.format(like_num)

resp = requests.get(url, cookies={'over18': '1'})
soup = BeautifulSoup(resp.text, 'lxml')

for item in soup.select('.r-ent'):
    title = item.select('.title')[0].text
    like = item.select('.nrec span')[0].text
    author = item.select('.author')[0].text
    date = item.select('.date')[0].text
    link = item.select('.title a')[0]['href']
    print('title = ', title)
    print('like = ', like)
    print('author = ', author)
    print('date = ', date)
    print('link = ', link)
    print('========')
```
