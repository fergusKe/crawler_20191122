```python
import requests
import time
import json
from bs4 import BeautifulSoup
```

```python
hostname = 'https://www.ptt.cc'
page = '/bbs/Gossiping/index.html'
url = hostname + page

resp = requests.get(url)
soup = BeautifulSoup(resp.text, 'lxml')
soup
```

# 帶入 cookie

```python
resp = requests.get(url, cookies={'over18': '1'})
soup = BeautifulSoup(resp.text, 'lxml')
soup
```

# 取出文章

```python
for item in soup.select('.r-ent'):
    title = item.select('.title')[0].text
    author = item.select('.author')[0].text
    date = item.select('.date')[0].text
    link = item.select('.title a')[0]['href']
    print('title = ', title)
    print('author = ', author)
    print('date = ', date)
    print('link = ', link)
    print('========')
```
